### Обработка пропусков

#### ceiling_height


```python
# напишем функцию для расчета пртопускв и их доли по отношению ко всему датафрейму:
def count_nan(column):
    count_nan = column.isna().sum()
    return count_nan
def count_dolya_nan(column):
    dolya=100*column.isna().mean()
    return dolya
```

```python
column = data['ceiling_height']  
print('Количество пропусков:', count_nan(column))        
print('Доля пропусков,%:', count_dolya_nan(column))
```

```python
# обработка пропусков
# Методом isna() найдём все строки с пропусками в столбцах и посчитаем строки с пропусками с помощью функции len().
print('Количество пропусков "ceiling_height":', len(data[data['ceiling_height'].isna()]))
#посчитаем Долю пропусков, %:
nan_ceiling_height = len(data[data['ceiling_height'].isna()])
data_length = len(data['total_images'])
dolya = ((nan_ceiling_height *100)/data_length)
#dolya= dolya.round(2)
print('Доля пропусков,%:', dolya)
```

Доля пропусков в категории "высота потолков" довольно-таки высока. По какой причине она не указана, сложно сказать. ЕСли это агенство - то возможно владелец не предоставил данную информацию. Заполнить чем-либо данные пропуски не представляетчя возможным. Тут нужно знать адреса домов, в которых представлены квартиры и попробовать соотнести их. Но этих данных у нас тоже нет. Поэтому оставим их пустыми.

#### floors_total

```python
print('Количество пропусков "floors_total":', len(data[data['floors_total'].isna()]))
#посчитаем Долю пропусков, %:
nan_floors_total = len(data[data['floors_total'].isna()])
dolya = ((nan_floors_total *100)/data_length)
#dolya= dolya.round(2)
print('Доля пропусков,%:', dolya)
```

Доля пропусков в категории "общее количество этажей в доме" очень мала: меньше 1%, всего 86 строчек из 23699. Заполнить их каким-то реальным значением на текущий момент не представляется возможным. Удалим эти строки. Влияния на остальные столбцы не произойдет.

```python
data = data.dropna(subset=['floors_total']).reset_index(drop=True)
print('Количество пропусков "floors_total":', print(data['floors_total'].isna().sum()))
```

#### living_area

Здесь доля пропусков также мала. Заполнить пропуски нечем. У нас имеются только три значения площадей: общая, жилая и кухня. Этого недостаточно для расчета жилой площади. Удалять пока эти пропуски мы не будем.

#### is_apartment

```python
print('Количество пропусков "is_apartment":', len(data[data['is_apartment'].isna()]))
#посчитаем Долю пропусков, %:
nan_is_apartment = len(data[data['is_apartment'].isna()])
dolya = ((nan_is_apartment *100)/data_length)
#dolya= dolya.round(2)
print('Доля пропусков,%:', dolya)
```

Доля пропусков категории, является ли недвижимость апартаменатми или нет, очень высока. Возможно все эти квартиры не являются апартаментами. Тогда можно присвоить значение False:

```python
# Для всех пар столбцов найдем коэффициент корреляции. Достаточно вызвать метод corr() без параметров.
df_final_pivot_data_reset.corr()
```

Если рост одной величины связан с уменьшением другой, коэффициент отрицательный.
